**1. Import the xml.etree.ElementTree module**


The first step in working with XML files is to import the xml.etree.ElementTree module. This module provides a simple way to parse XML data and create an object representation of the data.

````python
import xml.etree.ElementTree as ET
````

**2. Parse the XML file**

The next step is to parse the XML file using the ET.parse() method. This method takes the file path as an argument and returns an ElementTree object.

Download this XML file for practice:

[Link to Dataset](https://drive.google.com/file/d/1ExRvspSK4vD7qVtaJw_U8jt-UsNqzY2u/view?usp=sharing)


````python
tree = ET.parse('filepath')    #Paste the path to the xml file you downloaded
````

**3. Get the root element**

The ElementTree object returned by the ET.parse() method has a getroot() method that returns the root element of the XML file. We can use this method to get the root element.

In an XML file, the root element is the top-level element that contains all other elements. It is the starting point of the XML hierarchy and serves as the container for all the other elements in the document.

````python
root = tree.getroot()
print(root.tag)
````

**4. Accessing elements and attributes**


Once we have the root element, we can access its child elements and attributes using their tag names and attribute names. We can use the find() method to get the first occurrence of an element with a given tag name, or the findall() method to get all occurrences of an element with a given tag name. We can also access the attributes of an element using the dictionary-like attrib property.

````python
# access the elements and attributes
for book in root.iter('book'):
    title = book.find('title').text
    author = book.find('author').text
    year = int(book.find('year').text)
    publisher = book.find('publisher').text

    # do something with the data
    print(f'Title: {title}')
    print(f'Author: {author}')
    print(f'Year: {year}')
    print(f'Publisher: {publisher}\n')
````

In this example, we first parse the XML file using the ElementTree module and get the root element. We then use a for loop to iterate over all the book elements in the XML file, and use the find() and get() methods to access the elements and attributes. We convert the year value to an integer using the int() function. Finally, we print out the data in a formatted way.

Finally, we use the find() method again to get the title and author elements of the first book, and extract their text content using the text property of the Element object.

Note that if there are multiple book elements in the XML file, you can use the findall() method to get a list of all the book elements, and then iterate over the list to extract the desired information from each book.

**6. Accessing other Elements**

To access the details of the second book with id attribute equals to 2, you can use the find() method with an XPath expression that matches the book element with the id attribute equals to 2.

Here's an example code snippet that shows how to access the title and author of the second book with id attribute equals to 2:

````python
# find the book element with id attribute equals to 2
book2 = root.find("./book[@id='2']")

# check if book2 is None before accessing its attributes
if book2 is not None:
    title = book2.find('title').text
    author = book2.find('author').text
    print('Title:', title)
    print('Author:', author)
else:
    print('Book not found.')
````

In this example, we use the find() method with an XPath expression ./book[@id='2'] to find the book element with id attribute equals to 2. The . in the beginning of the expression means the current context node, which is the root element in this case. The [@id='2'] part of the expression is a condition that matches the book element with id attribute equals to 2. Then, we extract the title and author elements of the book and print their text content.
